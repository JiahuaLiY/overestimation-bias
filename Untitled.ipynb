```python
import gymnasium as gym
import numpy as np

from stable_baselines3 import DDPG, TD3
from stable_baselines3.common.noise import NormalActionNoise

from utils import QValuesCallback

import matplotlib.pyplot as plt
import pandas as pd

env = gym.make("LunarLander-v3",
               continuous=True,
               gravity=-10.0,
               enable_wind=False,
               wind_power=15.0,
               turbulence_power=1.5)

n_actions = env.action_space.shape[-1]

_action_noise = NormalActionNoise(mean=np.zeros(n_actions), sigma=0.1 * np.ones(n_actions))

_samplingSize = 800
_samplingFreq = 1000

_nbEpisodes = 100
_mcSeeds = [883, 368, 45, 747, 373, 523, 103, 705, 224, 424]

_total_timesteps = 300_000


def run(constructor: DDPG | TD3, seed: int):
    env.reset(seed=42)
    model = constructor("MlpPolicy",
                        env,
                        action_noise=_action_noise,
                        verbose=0,
                        learning_rate=1e-3,
                        buffer_size=200000,
                        learning_starts=10000,
                        gamma=0.98,
                        seed=seed,
                        policy_kwargs=dict(net_arch=[400, 300]))
    

    callback = QValuesCallback(samplingSize=_samplingSize,
                               samplingFreq=_samplingFreq,
                               nbEpisodes=_nbEpisodes,
                               mcSeeds=_mcSeeds,
                               gamma=.98,
                               mcEnv=env)
    model.learn(total_timesteps=_total_timesteps,
                callback=callback)
    return callback.qValues, callback.qValuesMC
```

```python
qValuesDDPG, qValuesDDPGMC = run(DDPG, 256)
```

```python
plt.plot(np.arange(0, _total_timesteps, _samplingFreq), qValuesDDPG)
plt.plot(np.arange(0, _total_timesteps, _samplingFreq), qValuesDDPGMC)
plt.show()
```

```python
plt.plot(np.arange(0, _total_timesteps, _samplingFreq*2), np.array(qValuesDDPG)[::2], "k.-")
plt.plot(np.arange(0, _total_timesteps, _samplingFreq*2), np.array(qValuesDDPGMC)[::2], "r.-")
plt.show()
```

```python
data = {
    "iter": np.arange(0, _total_timesteps, _samplingFreq),
    "q": qValuesDDPG,
    "qMC": qValuesDDPG
}

p = pd.DataFrame(data)
p.to_csv("mc1.csv")
```
